### 1. A simple loss fuction 
We want a parameter $\theta$ to produce an output close to a target value.

```python
# target value 
y_true = 10.0
# model parameter (what we will learn)
theta = 0.0
# input
x = 3.0
```

#### Model and loss:

```python
def model(x,theta):
    return theta*x
def loss(y_pred,y_true):
    return (y_pred-y_true)**2
```

### 2. Compute loss and slope (gradient)

```python
# forward pass
y_pred = model(x,theta)
current_loss = loss(y_pred,y_true)

print("Prediction:",y_pred)
print("Loss:",current_loss)
```

#### Gradient tell direction to move $\theta$

```python
def gradient(theta, x,y_true):
    y_pred = theta*x
    return 2*(y_pred - y_true)*2
```

### 3. One gradient descent step

```python
learning_rate = 0.01
grad = gradient(theta,x,y_true)
theta = theta-learning_rate*grad
print("Gradient:",grad)
print("Update theta:",theta)
```

#### Intuition
* Gradient sign → direction
* Learning rate → step size
* Update moves $\theta$ downhill

### 4. Watch gradient descent over iterations

```python
theta = 0.0
learning_rate = 0.01
for step in range(20):
    y_pred = model(x,theta)
    current_loss = loss(y_pred,y_true)
    grad = gradient(theta,x,y_true)
    
    theta -= learning_rate*grad
    print(f"Step {step:2d}| theta = {theta:.4f}|loss = {current_loss:.4f}")
```

#### What to observe
* Loss decreases steadily
* θ moves toward the best value
* Gradient magnitude shrinks near minimum

### 5. Effect of learning rate
Try changing:

```python
learning_rate = 0.1   # fast, may overshoot
learning_rate = 0.001 # slow, stable
```

#### Key insight
* Too large → jumps past minimum
* Too small → very slow learning
* Good α balances speed and stability

### 6. Visual intuition (optional but powerful)

```python
import numpy as np
import matplotlib.pyplot as plt
thetas = np.linspace(-1,5,100)
losses = [(model(x,t)-y_true)**2 for t in thetas]

plt.plot(thetas,losses)
plt.xlabel('theta')
plt.ylabel('loss')
plt.title("Loss landscape")
plt.show()
# Now mentally place each update step on this curve.
```

### 7. Core ML takeaway
* Loss tells how wrong you are
* Gradient tells which way to move
* Learning rate decides how far
* Update rule repeats until minimum is reached

This is the engine behind training neural networks.
