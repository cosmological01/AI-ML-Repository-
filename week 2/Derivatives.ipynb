### 1. Why derivatives matter in ML
Derivatives tell you:
* how fast loss changes
* which direction to move parameters
* how gradient descent works

In ML:
* derivative = gradient
* gradient = learning signal

### 2. Polynomial functions
Basic rule
$$\frac{d}{dx}(x^{n})=nx^{n-1}$$

#### Example 1: $x^{2}$

```python
# derivative of x^2
def d_x2(x):
    return 2*x
d_x2(3)
```

#### Example 2: quadratic loss 
$$L(w) = (w-5)^{2}$$

```python
def loss(w):
    return (w-5)**2
def grad_loss(w):
    return 2*(w-5)
grad_loss(3)
```

### ML meaning:
* negative gradient → increase w
* positive gradient → decrease w

### Exponential functions
Rule $\frac{d}{dx}(e^{x}) = e^{x}$ 

```python
import numpy as np
def exp_func(x):
    return np.exp(x)
def d_exp_func(x):
    return np.exp(x)
d_exp_func(0)

# Used in:
#     softmax
#     probability modeling
```
