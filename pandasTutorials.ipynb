# Exercise 2: Identifying and Removing Outliers

### Objective
Learn how to identify and remove outliers using quantiles and z-scores, and visualize the outliers using box plots.

### Steps
1. **Load the dataset**:
   - Use a dataset with numerical features (e.g., house prices, stock prices, or similar).
   
2. **Identify outliers**:
   - **Quantile-based method**: Calculate the lower and upper bounds using the 1st and 3rd quartiles (IQR method).
   - **Z-score method**: Use the standard deviation and mean to calculate Z-scores and identify points beyond a chosen threshold.

3. **Visualize the outliers**:
   - Use **box plots** to display the distribution of data and highlight the outliers.

4. **Remove the outliers**:
   - Filter the dataset to exclude outliers identified by either or both methods.
   - Create a cleaned version of the DataFrame.

5. **Compare before and after**:
   - Calculate summary statistics and visualize the data to compare the original and cleaned datasets.

### Notes
- Use libraries such as `pandas`, `numpy`, `matplotlib`, or `seaborn` for implementation.
- Experiment with different thresholds for Z-scores and quantiles to observe their effects.


```python
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import zscore
import matplotlib.pyplot as plt

diamonds_df = sns.load_dataset('diamonds')
diamonds_df.head()

# Step 2: Quantile-Based Outlier Removal
# Define Quantiles: For a chosen variable, such as price, calculate the 5th and 95th percentiles.
# Filter Data: Keep rows where price is within this range.

# Calculate the 5th and 95th percentiles for the 'price' column
q_low = diamonds_df['price'].quantile(0.05)
q_high = diamonds_df['price'].quantile(0.95)

# Filter the DataFrame based on quantiles
quantile_filtered_df = diamonds_df[(diamonds_df['price'] >= q_low) & (diamonds_df['price'] <= q_high)]
print("Quantile-based filtering - Rows before:", diamonds_df.shape[0], "Rows after:", quantile_filtered_df.shape[0])


# Step 3: Z-Score Outlier Removal
# Calculate Z-Scores: Use zscore() on numerical columns.
# Set Threshold: Keep rows where the z-score is within a defined threshold (e.g., below 3 standard deviations).

# Calculate z-scores for the 'price' column and filter based on a threshold
diamonds_df['z_score_price'] = zscore(diamonds_df['price'])
z_score_filtered_df = diamonds_df[(diamonds_df['z_score_price'].abs() <= 3)]
print("Z-score filtering - Rows before:", diamonds_df.shape[0], "Rows after:", z_score_filtered_df.shape[0])


# Step 4: Visualization Using Box Plots
# Visualize the outliers before and after removal with box plots.

# Set up figure and subplots for side-by-side comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Original data with potential outliers
sns.boxplot(data=diamonds_df, y='price', ax=axes[0])
axes[0].set_title("Original Data")

# Data after quantile-based filtering
sns.boxplot(data=quantile_filtered_df, y='price', ax=axes[1])
axes[1].set_title("Quantile-Based Outlier Removal")

# Data after z-score filtering
sns.boxplot(data=z_score_filtered_df, y='price', ax=axes[2])
axes[2].set_title("Z-Score Based Outlier Removal")

plt.tight_layout()
plt.show()
```

# Exercise 3: Outlier Detection and Removal

In this exercise, we will:

1. **Load a dataset with numerical features** (e.g., the Boston Housing dataset).
2. **Identify outliers** in the dataset using:
   - **Z-score**
   - **Interquartile Range (IQR)**
3. **Remove the outliers** and create a cleaned DataFrame.

### Instructions
- Use appropriate Python libraries such as `numpy`, `pandas`, or `scipy` for calculations.
- Visualize the outliers using plots (e.g., box plots or scatter plots).
- Ensure the cleaned DataFrame retains the integrity of the original dataset (except for the removed outliers).


```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy.stats import zscore

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target  # Add the target variable

print("Original dataset shape:", data.shape)

# Method 1: Outlier Detection using Z-score
def remove_outliers_zscore(df, threshold=3):
    z_scores = np.abs(zscore(df.iloc[:, :-1]))  # Exclude 'target' column
    return df[(z_scores < threshold).all(axis=1)]

data_zscore_cleaned = remove_outliers_zscore(data)
print("After Z-score cleaning:", data_zscore_cleaned.shape)

# Method 2: Outlier Detection using Interquartile Range (IQR)
def remove_outliers_iqr(df):
    Q1 = df.iloc[:, :-1].quantile(0.25)  # Exclude 'target' column
    Q3 = df.iloc[:, :-1].quantile(0.75)
    IQR = Q3 - Q1
    mask = ~((df.iloc[:, :-1] < (Q1 - 1.5 * IQR)) | (df.iloc[:, :-1] > (Q3 + 1.5 * IQR))).any(axis=1)
    return df[mask]

data_iqr_cleaned = remove_outliers_iqr(data)
print("After IQR cleaning:", data_iqr_cleaned.shape)

# Display the results
print("\nDataset shape comparison:")
print("Original:", data.shape)
print("Z-score cleaned:", data_zscore_cleaned.shape)
print("IQR cleaned:", data_iqr_cleaned.shape)
```

# Exercise 4: Missing Data Imputation

### Objective
Generate missing values in a dataset randomly and perform **imputation** for handling missing data.

### Steps:
1. **Generate Missing Values**:
   - Use `np.nan` to randomly introduce missing values in your dataset.

2. **Imputation Techniques**:
   - **For numerical columns**:
     - **Mean**: Fill missing values with the mean of the column.
     - **Median**: Fill missing values with the median of the column.
     - **Mode**: Fill missing values with the mode (most frequent value) of the column.

   - **For time-series data**:
     - **Forward Fill**: Fill missing values with the last known value.
     - **Backward Fill**: Fill missing values with the next available value.

### Notes:
- You can use `pandas` functions like `fillna()` for imputation and `numpy` for generating random `np.nan` values.
- Test the different imputation methods and see their effects on the dataset.


```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target  # Add the target variable

# Display the original data
print("Original data:")
# print(data.head())

# Task 1: Generate missing values randomly (use np.nan)
np.random.seed(42)  # For reproducibility
missing_rate = 0.1  # Set the missing data rate (10%)

# Randomly set missing values in numerical columns
num_missing = int(missing_rate * data.size)  # Number of missing values to introduce
missing_indices = np.random.choice(data.size, num_missing, replace=False)  # Random indices
data.values.ravel()[missing_indices] = np.nan  # Set these indices to NaN

print("\nData with missing values:")
# print(data.head())

# Task 2: Imputation

# 1. Impute with the mean (for numerical columns)
data_mean_imputed = data.copy()
data_mean_imputed.fillna(data.mean(), inplace=True)
print("\nData after Mean Imputation:")
# print(data_mean_imputed.head())

# 2. Impute with the median (for numerical columns)
data_median_imputed = data.copy()
data_median_imputed.fillna(data.median(), inplace=True)
print("\nData after Median Imputation:")
# print(data_median_imputed.head())

# 3. Impute with the mode (for numerical columns)
data_mode_imputed = data.copy()
data_mode_imputed.fillna(data.mode().iloc[0], inplace=True)  # Mode of each column
print("\nData after Mode Imputation:")
# print(data_mode_imputed.head())
```

# Task 8: Grouped Statistics

This task involves grouping data by a categorical feature and calculating various aggregated metrics such as:

- **Mean**: The average value of the group.
- **Median**: The middle value when the data is sorted.
- **Standard Deviation**: The measure of the spread of data within the group.
- **Count**: The number of data points in the group.

## Steps to Perform:

1. Import the necessary libraries (e.g., `pandas`).
2. Load the dataset containing both categorical and numerical features.
3. Group the dataset by the categorical feature.
4. Apply aggregation functions like `mean`, `median`, `std`, and `count` to the grouped data.
5. Display the results in a structured format (e.g., a DataFrame).

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Group data by species
grouped = data.groupby('species')

# Calculate aggregated statistics
aggregated_stats = grouped.agg(
    mean_sepal_length=('sepal length (cm)', 'mean'),
    median_sepal_length=('sepal length (cm)', 'median'),
    std_sepal_length=('sepal length (cm)', 'std'),
    count_sepal_length=('sepal length (cm)', 'count'),
    mean_sepal_width=('sepal width (cm)', 'mean'),
    median_sepal_width=('sepal width (cm)', 'median'),
    std_sepal_width=('sepal width (cm)', 'std')
)

# Display the aggregated statistics
print(aggregated_stats)
```

## Task 9: Mutual Information
Mutual information measures the dependency between each feature and the target variable. It is a non-negative value, with higher values indicating a stronger relationship.

### Steps:
1. Import the necessary libraries and the dataset.
2. Use the `mutual_info_classif` or `mutual_info_regression` function from `sklearn.feature_selection` (depending on whether the target is categorical or continuous).
3. Compute mutual information scores for all features with respect to the target variable.
4. Rank the features by their importance.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import mutual_info_classif, VarianceThreshold
import pandas as pd

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target  # Target variable

# Task 9: Mutual Information
# Calculate mutual information between each feature and the target variable
mi_scores = mutual_info_classif(X, y)
mi_scores_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

print("Features ranked by mutual information with the target:")
print(mi_scores_series)

# Task 10: Variance Threshold
# Identify and drop low-variance features
variance_threshold = 0.01  # Set the variance threshold
selector = VarianceThreshold(threshold=variance_threshold)
X_high_variance = selector.fit_transform(X)

# Get the names of retained features
retained_feature_mask = selector.get_support()  # Boolean mask of retained features
retained_features = X.columns[retained_feature_mask]

print("\nRetained features after applying variance threshold:")
print(retained_features)

# Optional: Create a DataFrame with only the retained features
X_reduced = X[retained_features]
```

Normalization rescales the values of numerical features to the range [0, 1]. This is useful when the features have varying scales. 
## Task 12: Log Transformation
Apply log transformations to skewed features.
Analyze the impact on feature distribution.

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Task 11: Normalization and Scaling
# Normalize numerical features to [0, 1]
scaler_minmax = MinMaxScaler()
normalized_data = pd.DataFrame(scaler_minmax.fit_transform(data), columns=data.columns)

# Standardize features to have mean 0 and standard deviation 1
scaler_standard = StandardScaler()
standardized_data = pd.DataFrame(scaler_standard.fit_transform(data), columns=data.columns)

# Task 12: Log Transformation
# Apply log transformation to skewed features
log_transformed_data = data.copy()
for col in log_transformed_data.columns:
    # Adding a small constant to avoid log(0)
    log_transformed_data[col] = np.log1p(log_transformed_data[col])

# Analyze the impact on feature distribution
# Plotting original vs transformed data for the first feature
feature = data.columns[0]
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(data[feature], bins=20, color='blue', alpha=0.7, label='Original')
plt.title(f'Original: {feature}')
plt.legend()

plt.subplot(2, 2, 2)
plt.hist(normalized_data[feature], bins=20, color='green', alpha=0.7, label='Normalized')
plt.title(f'Normalized: {feature}')
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(standardized_data[feature], bins=20, color='orange', alpha=0.7, label='Standardized')
plt.title(f'Standardized: {feature}')
plt.legend()

plt.subplot(2, 2, 4)
plt.hist(log_transformed_data[feature], bins=20, color='red', alpha=0.7, label='Log Transformed')
plt.title(f'Log Transformed: {feature}')
plt.legend()

plt.tight_layout()
plt.show()
```

# Task 14: Efficient Joins

In this notebook, we will explore efficient methods to join two large DataFrames using different join strategies (`inner`, `outer`, `left`, `right`) with the `pd.merge()` function.

---

```python
from sklearn.datasets import load_iris, load_wine
import pandas as pd

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['id'] = range(1, len(iris_df) + 1)  # Add a unique identifier column

# Load the wine dataset
wine = load_wine()
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['id'] = range(50, 50 + len(wine_df))  # Add a unique identifier column with some overlap

# Perform optimized merges
# Inner join: Only common keys between the two DataFrames
inner_join = pd.merge(iris_df, wine_df, on='id', how='inner')

# Outer join: All keys from both DataFrames
outer_join = pd.merge(iris_df, wine_df, on='id', how='outer')

# Left join: All keys from the first (left) DataFrame
left_join = pd.merge(iris_df, wine_df, on='id', how='left')

# Right join: All keys from the second (right) DataFrame
right_join = pd.merge(iris_df, wine_df, on='id', how='right')

# Display the results
print("Inner Join Result:")
print(inner_join.head(), '\n')

print("Outer Join Result:")
print(outer_join.head(), '\n')

print("Left Join Result:")
print(left_join.head(), '\n')

print("Right Join Result:")
print(right_join.head(), '\n')
```

### Task 17: Lag Features for Time Series
We’ll simulate a time-series dataset for this task.

```python
import pandas as pd
import numpy as np

# Generate a synthetic time series dataset
np.random.seed(42)
date_range = pd.date_range(start="2020-01-01", periods=100, freq="D")
data = pd.DataFrame({'date': date_range, 'value': np.random.randn(100).cumsum()})

# Create lag features
data['lag_1'] = data['value'].shift(1)  # 1-day lag
data['lag_2'] = data['value'].shift(2)  # 2-day lag

# Drop NaN values introduced by lagging
data = data.dropna()

# Display the resulting DataFrame
print(data.head())

# Plot the original and lagged series for visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['value'], label='Original', color='blue')
plt.plot(data['date'], data['lag_1'], label='Lag 1', color='orange', linestyle='--')
plt.plot(data['date'], data['lag_2'], label='Lag 2', color='green', linestyle='--')
plt.legend()
plt.title("Time Series with Lag Features")
plt.show()
```

### Task 18: Rolling Statistics
We’ll calculate moving averages, rolling standard deviations, and cumulative sums.

```python
# Calculate rolling statistics
data['rolling_mean_5'] = data['value'].rolling(window=5).mean()
data['rolling_std_5'] = data['value'].rolling(window=5).std()
data['cumulative_sum'] = data['value'].cumsum()

# Display the resulting DataFrame
print(data[['date', 'value', 'rolling_mean_5', 'rolling_std_5', 'cumulative_sum']].head(10))

# Plot rolling statistics
plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['value'], label='Original', color='blue')
plt.plot(data['date'], data['rolling_mean_5'], label='Rolling Mean (5)', color='red', linestyle='--')
plt.fill_between(data['date'], 
                 data['rolling_mean_5'] - data['rolling_std_5'], 
                 data['rolling_mean_5'] + data['rolling_std_5'], 
                 color='gray', alpha=0.2, label='Rolling Std Dev (5)')
plt.legend()
plt.title("Rolling Statistics")
plt.show()
```

### Task 19: Feature Binning
Using the California Housing dataset for numerical feature binning.

```python
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt

# Load the Wine dataset
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target  # Add target as a categorical column

# Bin the 'alcohol' feature into intervals
bins = [11, 12, 13, 14, 15]  # Define bin edges for alcohol content
labels = ['Low', 'Medium', 'High', 'Very High']  # Define bin labels
df['Alcohol_Binned'] = pd.cut(df['alcohol'], bins=bins, labels=labels, include_lowest=True)

# Analyze the impact of binning on feature distribution
print(df[['alcohol', 'Alcohol_Binned']].head(10))
df['Alcohol_Binned'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Binned Alcohol Content Distribution')
plt.xlabel('Alcohol Bins')
plt.ylabel('Count')
plt.show()

# Correlation between alcohol bins and target variable
print(df.groupby('Alcohol_Binned')['target'].value_counts(normalize=True))
```
