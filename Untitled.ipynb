```python
def prime_check(n):
    lists = list(range(n+1))
    remdr = [n%num for num in lists[2:n]]
    if 0 in remdr: return False
    else: return True
```

```python
def all_prime(start, end):
    adx = []
    for i in range(start+1,end+1):
        if prime_check(i)==True:
            adx.append(i)
    return adx
```

```python
def freq_char(text):
    add={char: text.count(char) for char in set(text)}
    return add

freq_char("banana")
```

```python
def palindrome(text):
    text = text.replace(" ","")
    text = text.lower()
    rev_text = list(text)[::-1]
    if list(text)==rev_text:
        return print('palendrome')
    else: print('Not palendrome')
```
